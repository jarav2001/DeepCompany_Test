```python
csv_file = Path("tb_product_202410181030.csv")
```

```python
df = pd.read_csv(csv_file,
        header=0,
        index_col= ["id"],
        low_memory = False,         
        usecols=["id","id_trademark","id_measurement_unit","id_record_status","id_company_variation","id_product_category","code","name","registration_date","description","model","manufacturer","keyword","id_subscription","id_product_category_l1","sku"], 
        sep = ',')
```

```python
print (df)
```

```python
df.dtypes
```

```python
df_na = df.dropna()
```

```python
print(df_na)
```

```python
grouped_data = df_na.groupby('manufacturer')
```

```python
grouped_data.describe()
```

```python
count_data = df_na.groupby('manufacturer')['name'].nunique()
```

```python
count_data_mayores = count_data.sort_values(ascending=False)
```

```python
print(count_data_mayores.head(50))
```

```python
count_data_mayores.head(30).plot(kind='bar')
```
